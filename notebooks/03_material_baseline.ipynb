# Simple baseline (Material)

This notebook improves upon the previous baseline by adding material count as a feature and improving code quality. To recap, we were able to achieve a 53% accuracy using only Elo as a feature.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import duckdb # we rely on duckdb for parsing movedata
duckdb.sql("LOAD aixchess")
```

```python
data_path = "subset.parquet
df = pd.read_parquet(data_path)
df.head()
```

```python

games = duckdb.read_parquet("subset.parquet")

game_positions = duckdb.sql("""
SELECT
  lc.* EXCLUDE (movedata, clocks_white, clocks_black, tournament),
  t.ply,
  UNNEST(board_at_position(lc.movedata, t.ply))
FROM games AS lc
CROSS JOIN LATERAL (
  SELECT 1 + CAST(floor(random() * lc.ply_count) AS INTEGER) AS ply
) AS t
""")
game_positions.df()
```

```python
material_counts = duckdb.sql("""SELECT 
    lichess_id,
    list_sum(list_transform(pieces, x -> CASE x
        WHEN 'p' THEN 1 WHEN 'n' THEN 3 WHEN 'b' THEN 3 
        WHEN 'r' THEN 5 WHEN 'q' THEN 9 ELSE 0 END
    )) AS white_material,
    list_sum(list_transform(pieces, x -> CASE x
        WHEN 'P' THEN 1 WHEN 'N' THEN 3 WHEN 'B' THEN 3 
        WHEN 'R' THEN 5 WHEN 'Q' THEN 9 ELSE 0 END
    )) AS black_material
FROM (
    SELECT lichess_id, list_value(UNPACK(COLUMNS('[a-h][1-8]'))) AS pieces
    FROM game_positions
)
""")
material_counts.df()
```

```python
tbl = duckdb.sql("""SELECT gp.lichess_id, ply_count, white_rating, black_rating, result, white_material, black_material 
FROM game_positions gp 
JOIN material_counts mc ON mc.lichess_id = gp.lichess_id""")
df = tbl.df()
```

```python
# Select out the features and labels
labels = 'result'
features = ['white_rating', 'black_rating', 'white_material', 'black_material']

X_df = df[features]
y_df = df[labels]
X_df, y_df
```

```python
# Note: this function has changed from the previous notebook
def preprocess(X_df, y_df, test_size=0.2, random_state=None):
    # removed preprocessing as it showed no benefit for Logistic Regression
    X, y = X_df.to_numpy(), y_df.to_numpy()
    
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

def train_baseline(X_train, y_train, max_iter=10000) -> LogisticRegression:
    model = LogisticRegression(max_iter=max_iter)
    model.fit(X_train, y_train)
    return model

def eval_model(model, X_test, y_test):
    y_preds = model.predict(X_test)
    num_correct = (y_preds == y_test).sum()
    total = y_test.shape[0]
    print(f"Accuracy: {num_correct}/{total} = {(num_correct / total):.3f}")
    return dict(num_correct=num_correct, total=total, accuracy=num_correct / total)

```

```python
X_train, X_test, y_train, y_test = preprocess(X_df, y_df, random_state=0)
model = train_baseline(X_train, y_train)
_ = eval_model(model, X_test, y_test)
```

### Accuracy jump. What's next?

We've improved accuracy a fair bit just by adding material (53% -> 58%). The next logical step is to start factoring in the actual positions 
of the pieces -- this should straightforwardly improve performance even on a linear model -- a knight in the center is worth more than one 
on the edge.

We can also add whose turn it is, by calculating whether the ply is odd (white) or even (black).

```python
# Not a very impressive model.
# Let's do a couple of small modifications, such as normalizing the elo scores beforehand.

print(f"mean white (rating, stddev): {(X_df['white_rating'].mean(), X_df['white_rating'].std())}")

def process_elo(ratings: pd.Series):
    return (ratings - ratings.mean()) / ratings.std()

_X_norm = pd.DataFrame({col: process_elo(X_df[col]) for col in X_df.columns})
print(f"mean white (rating, stddev): {(_X_norm['white_rating'].mean(), _X_norm['white_rating'].std())}")
```

```python
# Looks right. Let's create a reusable data processing function, that returns our numpy arrays ready for training.

def preprocess(X_df, y_df, test_size=0.2, random_state=None):
    X_df = pd.DataFrame({col: process_elo(X_df[col]) for col in X_df.columns})
    X, y = X_df.to_numpy(), y_df.to_numpy()
    
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

# Let's create some for training and evaluating a baseline, too.

def train_baseline(X_train, y_train, max_iter=10000) -> LogisticRegression:
    model = LogisticRegression(max_iter=max_iter)
    model.fit(X_train, y_train)
    return model

def eval_model(model, X_test, y_test):
    y_preds = model.predict(X_test)
    num_correct = (y_preds == y_test).sum()
    total = y_test.shape[0]
    print(f"Accuracy: {num_correct}/{total} = {(num_correct / total):.3f}")
    return dict(num_correct=num_correct, total=total, accuracy=num_correct / total)
```

```python
X_train, X_test, y_train, y_test = preprocess(X_df, y_df, random_state=0)
model = train_baseline(X_train, y_train)
_ = eval_model(model, X_test, y_test)
```

```python
# It's actually identical -- normalizaion doesn't do anything. 
# I guess that makes sense, because there's no activation, and it's probably doing a direct solve.
# Let's try changing the random seed, and seeing what some typical scores are.
def benchmark(seeds=range(10)):
    for seed in seeds:
        X_train, X_test, y_train, y_test = preprocess(X_df, y_df, random_state=seed)
        model = train_baseline(X_train, y_train)
        print(f"(seed = {seed})", end=' ')
        _ = eval_model(model, X_test, y_test)

benchmark()
```

```python
# Lucky fluke? Let's check some more.
benchmark(range(50))
```

```python
# Our first was the best. Strange. Anyway.
# Let's try a heuristic function:
class HeuristicModel():
    def predict(self, X):
        outcomes = np.array(['1-0', '0-1'])

        return outcomes[np.argmax(X + np.array([0.04, 0]), axis=1)]

_ = eval_model(HeuristicModel(), X_test, y_test)
```
